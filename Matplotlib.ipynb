# Matplotlib

Matplotlib is a Python 2D plotting library which produces publication quality figures.

The objects in matplotlib:

- a container for all plots was created (returned as a **Figure** object)
- a container for the plot was positioned on a grid (the plot returned as an **Axes** object)
- visual symbols were added to the plot (using the Axes methods)

## 1 Line chart plot

```python
import matplotlib
import matplotlib.pyplot as plt
# magic functions
%matplotlib inline

x1 = ['1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000']
y1 = [1, 3, 4, 4, 5, 2, 6, 8]
x2 = ['1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000']
y2 = [1, 5, 4, 6, 1, 7, 6, 8]
```

```python
# single plot

plt.plot(x1, y1, c='red')  # any iterable object

plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Year Trends')
```

```python
# multiple lines in the same plot
# share same x axes

fig = plt.figure(figsize=(6, 3))

plt.plot(x1, y1, c='red', label='year1')
plt.plot(x2, y2, c='blue', label='year2')

plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Year Trends')
plt.legend(loc='upper left')
```

```python
# multiple plots

fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x1, y1, c='red', label='year1')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number')
ax1.set_title('Year Trends 1')
ax1.legend(loc='upper left')
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

ax2.plot(x2, y2, c='blue', label='year2')
ax2.set_xlabel('Year')
ax2.set_ylabel('Number')
ax2.set_title('Year Trends 2')
ax2.legend(loc='upper left')
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
```

## 2 Some other plots

```python
# bar plots
# no order for the x axis

import numpy as np

bar_positions = np.arange(8) + 0.75
tick_positions = range(1, 9)

fig, ax = plt.subplots()
ax.bar(bar_positions, y1, width = 0.5)  # y

ax.set_xticks(tick_positions)
ax.set_xticklabels(x1, rotation=90)  # x
ax.set_xlabel('Year')
ax.set_ylabel('Number')
ax.set_title('Year Trends')
```

```python
# scatter plots
# a plot containing just the markers
# a scatter plot helps us determine if 2 columns are weakly or strongly correlated

fig, ax = plt.subplots()
ax.scatter(x1, y1)

ax.set_xlabel('Year')
ax.set_ylabel('Number')
ax.set_title('Year Trends')
```

```python
# histograms

z = [1,2,3,3,5,6,7,2,4,5]

fig,ax = plt.subplots()
ax.hist(z, 20, range=(0,10))
ax.set_xlabel('Number')
```

```python
# box plots

z = [1,2,3,3,5,6,7,2,4,5]

fig, ax = plt.subplots()
ax.boxplot(z)
ax.set_xlabel('Number')
```

## 3 Plot aesthetics

```python
# use RGB color
# change the line width
plt.plot(x1, y1, c=(0 / 255, 107 / 255, 164 / 255), linewidth=2)

# add lable
plt.xlabel('Year')
plt.ylabel('Number')
# add title
plt.title('Year Trends')
# rotate tick
plt.xticks(rotation=90)
# set range
plt.ylim(0, 10)
# reomove tick mark
plt.tick_params(bottom="off", top="off", left="off", right="off")
# remove spine
plt.gca().spines["bottom"].set_visible(False)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["left"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
# add text annotations
plt.text(1999, 8, "year1")
```
